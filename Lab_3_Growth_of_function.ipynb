# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
import math
import matplotlib.pyplot as plt

def log_linear_example(n):
    steps = 0
    for i in range(n):
        steps += math.log2(n)
    return steps

input_sizes = [10, 100, 1000, 5000, 10000]
results = [log_linear_example(n) for n in input_sizes]

plt.figure(figsize=(8,5))
plt.plot(input_sizes, results, marker='o', label="O(n log n)")
plt.plot(input_sizes, input_sizes, marker='s', label="O(n) Linear")
plt.plot(input_sizes, [math.log2(n) for n in input_sizes], marker='^', label="O(log n)")

plt.title("Growth of O(n log n) vs O(n) vs O(log n)")
plt.xlabel("Input Size (n)")
plt.ylabel("Steps (approx)")
plt.legend()
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
import matplotlib.pyplot as plt
import numpy as np

def polynomial_example(n):
    steps = 0
    for i in range(n):
        for j in range(n):
            steps += 1
    return steps

n_values = list(range(1, 51))  # Using a smaller range for n due to n^2 complexity
time_values = []

for n in n_values:
    time_values.append(polynomial_example(n))

plt.figure()
plt.plot(n_values, time_values, label="O(n^2) Polynomial")
plt.xlabel("Input size n")
plt.ylabel("Time (arbitrary units)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.legend()
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
import matplotlib.pyplot as plt
import numpy as np

def exponential_example(n):
    # This function simulates exponential growth, e.g., 2^n operations
    # For practical plotting purposes, we'll return 2^n directly as the 'time' or 'steps'
    return 2**n

# Define a range for n. Exponential functions grow very rapidly,
# so a small range for n is necessary to keep values manageable for plotting.
n_values = list(range(1, 11))  # From 1 to 10
time_values = []

for n in n_values:
    time_values.append(exponential_example(n))

plt.figure()
plt.plot(n_values, time_values, label="O(c^n) Exponential")
plt.xlabel("Input size n")
plt.ylabel("Time (arbitrary units)")
plt.title("O(c^n) Exponential Time Complexity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define functions for each time complexity
def o_1(n): # Constant
    return np.ones_like(n)

def o_log_n(n): # Logarithmic
    # Handle n=0 case to avoid log(0) error, using a small epsilon if n can be 0
    return np.log2(np.maximum(n, 1))

def o_n(n): # Linear
    return n

def o_n_log_n(n): # Log-Linear
    # Handle n=0 case
    return n * np.log2(np.maximum(n, 1))

def o_n_squared(n): # Polynomial (n^2)
    return n**2

def o_2_power_n(n): # Exponential (2^n)
    return 2**n

# Define a range of input sizes
n_values = np.arange(1, 21) # From 1 to 20 for better visualization of exponential growth

# Calculate time values for each complexity
time_o_1 = o_1(n_values)
time_o_log_n = o_log_n(n_values)
time_o_n = o_n(n_values)
time_o_n_log_n = o_n_log_n(n_values)
time_o_n_squared = o_n_squared(n_values)

# For exponential, use a smaller range of n or scale values for plotting if n is large
# Here, we'll plot it on the same graph, understanding it will quickly dominate
time_o_2_power_n = o_2_power_n(n_values)

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(n_values, time_o_1, label='O(1) Constant', marker='o', markersize=4)
plt.plot(n_values, time_o_log_n, label='O(log n) Logarithmic', marker='o', markersize=4)
plt.plot(n_values, time_o_n, label='O(n) Linear', marker='o', markersize=4)
plt.plot(n_values, time_o_n_log_n, label='O(n log n) Log-Linear', marker='o', markersize=4)
plt.plot(n_values, time_o_n_squared, label='O(n^2) Polynomial', marker='o', markersize=4)

# Plot exponential only for a very small range, otherwise it makes other curves flat
# For better visualization, we might limit the n_values for exponential or adjust y-axis.
# For this example, let's plot it but acknowledge its rapid growth.
plt.plot(n_values, time_o_2_power_n, label='O(2^n) Exponential', marker='o', markersize=4, linestyle='--')

plt.xlabel("Input size (n)")
plt.ylabel("Operations / Time (arbitrary units)")
plt.title("Comparison of Time Complexities")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(0, 100) # Set a y-axis limit to better see the lower complexities
plt.xticks(n_values)
plt.show()
```
